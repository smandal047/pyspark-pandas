### Import data/csv and Output sample csv

```python
file = '../data/accepted_2007_to_2018Q4.csv'
file_type = file.split('.')[-1]

options = {
  'inferSchema': True,
  'header': True,
  'sep': ','
}

df = spark.read.format(file_type).options(**options).load(file)
```

```python
# df.limit(10).write.format("csv").mode("overwrite").saveAsTable('LCD_Analysis_4excel')
filepath = '../output/LCD_Analysis_4excel.csv'
df.limit(10).toPandas().to_csv(filepath)
```

```python
df = df.select('loan_amnt', 'funded_amnt', 'funded_amnt_inv', 'term', 'int_rate', 'installment', 'sub_grade', 'emp_title',\
               'emp_length', 'home_ownership', 'annual_inc', 'verification_status', 'loan_status', 'pymnt_plan', 'purpose',\
               'addr_state', 'dti', 'delinq_2yrs', 'fico_range_low', 'fico_range_high', 'inq_last_6mths', 'mths_since_last_delinq',\
               'open_acc', 'pub_rec', 'revol_bal', 'revol_util', 'total_acc', 'total_pymnt', 'total_pymnt_inv', 'total_rec_prncp',\
               'last_fico_range_high', 'last_fico_range_low', 'application_type', 'annual_inc_joint', 'dti_joint', 'verification_status_joint',\
               'acc_now_delinq', 'tot_coll_amt', 'tot_cur_bal', 'total_bal_il', 'il_util', 'max_bal_bc', 'all_util', 'total_rev_hi_lim',\
               'inq_fi', 'total_cu_tl', 'avg_cur_bal', 'bc_open_to_buy', 'bc_util', 'delinq_amnt', 'mort_acc', 'tot_hi_cred_lim',\
               'total_bal_ex_mort', 'total_bc_limit', 'total_il_high_credit_limit')
display(df)
```

### pre-processing the columns

#### Part 1: Remove Text

```python
from pyspark.sql.functions import regexp_extract, col
```

```python
df = df.withColumn('term_months', regexp_extract(col('term'), r'\d+', 0))
display(df)
```

```python
df.select('term', 'term_months').show(10)
```

```python
df = df.withColumn('emp_length_years', regexp_extract(col('emp_length'),r'\d+', 0))
display(df)
```

```python
df.select('emp_length', 'emp_length_years').show()
```

```python
df.select('dti', 'dti_joint').show(10)
```

```python
from pyspark.sql.functions import coalesce, isnull
```

```python
df.select('dti', 'dti_joint', coalesce(col('dti_joint'), col('dti'))).show(10)
```

```python
df = df.withColumn('dti_joint_merged', coalesce(col('dti_joint'), col('dti')))
display(df)
```
