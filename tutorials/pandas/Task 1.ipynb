```python
import pandas as pd
import numpy as np
```

***read csv***

```python
df = pd.read_csv('../csv_data/pre_process_1.csv')
df
```

***Create a new column of 'personID' and save it to a file***

```python
df.info()
```

```python
personID = pd.DataFrame({'PersonID': range(1,11)})
personID
```

```python
df = pd.concat([personID, df], axis=1)
df
```

***Add new rows in the data-frame***

```python
new_data = pd.DataFrame({'PersonID': [11, 12, 13],
                      'Country': ['India', 'Australia', 'France'],
                      'Age': ['NaN', 55, 82],
                      'Salary':['NaN', 46000, 30000],
                      'Purchased': ['Yes', 'Yes', 'Yes',]
                  })
new_data
```

```python
df = pd.concat([df, new_data])
df = df.reset_index().drop('index', axis=1)
df
```

***creating a new dataframe for merging***

```python
df1 = pd.DataFrame({
    'ID': range(1, 14),
    'Height': 5 + np.random.sample(13)*5,
    'Weight': 30 + np.random.sample(13)*10
})
df1
```

***merge df and df1***

```python
final_df = df.merge(df1, left_on='PersonID', right_on='ID').drop('ID', axis=1)
final_df
```

***using melt function to melt down columns (not included in final_df)***

```python
final_df.melt(id_vars=['PersonID','Country','Age','Salary','Height','Weight'], var_name='Ori_Purchased', value_name='Purchased')
```

```python
replace_purchased = final_df.Purchased.apply(lambda x : True if x=='Yes' else False)
final_df = final_df.assign(Purchased=replace_purchased)
final_df
```

***save to file***

```python
final_df.to_csv('../csv_output/pre_process_1.csv', index=False)
pd.read_csv('../csv_output/pre_process_1.csv')
```
